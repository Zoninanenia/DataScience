```python
import pandas as pd
import seaborn as sns
```

# 1.

```python
players = pd.read_csv('players.csv')
players
```

```python
players = players.iloc[ : , :8]
players
```

```python
players[['shots', 'passes', 'tackles', 'saves']] = players.loc[:, ['shots', 'passes', 'tackles', 'saves']].div(players['minutes'], axis=0)
players
```

```python
players['position'] = pd.Categorical(players.position, categories = ['goalkeeper', 'defender', 'midfielder', 'forward'], ordered = True)
players.info()
```

```python
players = players.sort_values('position')
players
```

# 2.

```python
sns.displot(players.passes, kind='kde', height=3, aspect=1.5)
```

# 3.

```python
sns.displot(players[players['team'] == 'USA'].shots, kind='hist', height=3)
```

# 4.

```python
sns.relplot(players, x='passes', y='shots', height=3, aspect=1)
```

# 5.

```python
sns.relplot(players[players['team'].isin(['Brazil', 'Spain', 'USA'])], x='passes', y='shots', hue='team', style='position', height=3, aspect=1)
```

# 6.

```python
sns.relplot(players[players['team'].isin(['Brazil', 'Spain', 'USA'])], x='tackles', y='shots', col='team', row='position', height=3, aspect=1)
```

# 7.

```python
sns.catplot(players[players['team'].isin(['Brazil', 'Spain', 'USA'])], x='team', y='passes', kind='box', hue='team', height=4, aspect=1)
```

# 8.

```python
sns.catplot(players[players['team'].isin(['Brazil', 'Spain', 'USA'])], x='team', y='passes', hue='position', kind='swarm', height=4, aspect=1)
```

# 9.

```python
sns.catplot(players[players['team'].isin(['Brazil', 'Spain', 'USA'])], x='position', y='tackles', kind='bar', estimator='mean', hue='team', height=4, aspect=1)
```

# 10.

```python
filtered_teams = ['Brazil', 'Japan', 'USA', 'Spain', 'Algeria', 'France', 'England', 'South Korea']
team_data = players[players['team'].isin(filtered_teams)].groupby('team')[['passes', 'shots']].sum()

sns.relplot(team_data, x='passes', y='shots', hue='team', height=4, aspect=1)
```
