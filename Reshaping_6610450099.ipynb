```python
drinks.rename(columns={'country': 'Country'}, inplace=True)
drinks.set_index('Country', inplace=True)
medals.set_index('Country', inplace=True)

medals = medals.join(drinks[['continent']], how='left')
medals
```

# 8.

```python
medals['continent'] = medals['continent'].fillna('OT');
medals.loc[medals.index == 'Australasia', 'continent'] = 'OC';
medals.loc[medals.index.isin(['Canada', 'United States', 'Mexico']), 'continent'] = 'NA';
medals.loc[medals.index.isin(['Cuba', 'Jamaica']), 'continent'] = 'SA';
medals.loc[medals.index.isin(['Czechoslovakia', 'Yugoslavia', 'United Team of Germany', 'East Germany', 'West Germany', 'Great Britain', 'Russia', 'Russian Empire', 'Soviet Union']), 'continent'] = 'EU'
wide_medals = medals
wide_medals
```

# 9.

```python
medals.reset_index(drop=False, inplace=True)
medals = medals.set_index(['continent', 'Country'])
medals
```

# 10.

```python
medals.sort_index(ascending=[False, True])
```

# 11.

```python
x = pd.MultiIndex.from_tuples([
    ('', 'Code'),
    ('Summer', 'SummerGame'),
    ('Summer', 'SummerGold'),
    ('Summer', 'SummerSilver'),
    ('Summer', 'SummerBronze'),
    ('Winter', 'WinterGame'),
    ('Winter', 'WinterGold'),
    ('Winter', 'WinterSilver'),
    ('Winter', 'WinterBronze'),
])

medals.columns = x
medals
```

# 12.

```python
y = medals.columns.swaplevel()
medals.columns = y
medals
```

# 13.

```python
id_vars = ['continent', 'Country']
value_vars = ['SummerGold', 'SummerSilver', 'SummerBronze', 'WinterGold', 'WinterSilver', 'WinterBronze']

long_medals = pd.melt(wide_medals, id_vars=id_vars + ['SummerGame', 'WinterGame'],
                      value_vars=value_vars, var_name='Medal_Season', value_name='Count')

long_medals['Season'] = long_medals['Medal_Season'].str.extract('(Summer|Winter)')
long_medals['Medal'] = long_medals['Medal_Season'].str.extract('(Gold|Silver|Bronze)')

long_medals['Game'] = long_medals.apply(
    lambda row: row['SummerGame'] if row['Season'] == 'Summer' else row['WinterGame'], axis=1
)

long_medals.drop(columns=['Medal_Season', 'SummerGame', 'WinterGame'], inplace=True)
long_medals = long_medals[['continent', 'Country', 'Season', 'Game', 'Medal', 'Count']]
long_medals = long_medals.sort_values(by='Medal')

long_medals
```
