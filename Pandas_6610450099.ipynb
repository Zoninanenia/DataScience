# 22 ประเทศที่ได้เหรียญรวมทุกประเภทของฤดูหนาวมากกว่าเหรียญรวมฤดูร้อน ให้คอลัมของตารางแสดงชื่อประเทศ เหรียญรวมฤดูร้อน เหรียญรวมฤดูหนาว

```python
medals['summer_total'] = medals['SummerGold'] + medals['SummerSilver'] + medals['SummerBronze']
medals['winter_total'] = medals['WinterGold'] + medals['WinterSilver'] + medals['WinterBronze']

medals[(medals.winter_total > medals.summer_total)][['Country', 'summer_total', 'winter_total']]
```

# 23 ประเทศที่ไม่เคยได้เหรียญฤดูร้อน แต่ได้เหรียญฤดูหนาว หรือประเทศที่ได้แต่เหรียญฤดูร้อนไม่เคยได้เหรียญฤดูหนาว

```python
medals[((medals.summer_total == 0) & (medals.winter_total > 0)) | ((medals.summer_total > 0) & (medals.winter_total == 0))][['Country']]
```

# 24 แสดงประเทศเรียงตามจำนวนเหรียญทองฤดูร้อนจากน้อยไปมาก

# 25 แสดงประเทศเรียงตามจำนวนเหรียญทองฤดูร้อนจากมากไปน้อย ถ้าได้เหรียญทองฤดูร้อนเท่ากันให้เรียงตามจน.เข้าแข่งของฤดูร้อนจากน้อยไปมาก

```python
medals.sort_values(['SummerGold', 'SummerGame'], ascending=[False, True])
```

# 26 สร้างคอลัมใหม่ชื่อ 'score' คำนวณคะแนนรวม โดยให้เรียญทองได้ 3 เหรียญเงินได้ 2 เหรียญทองแดงได้ 1 แสดงข้อมูล 10 ประเทศแรกที่ได้คะแนนเยอะสุด

```python
medals['score'] = (((medals.SummerGold + medals.WinterGold) * 3)
                + ((medals.SummerSilver + medals.WinterSilver) * 2)
                + ((medals.SummerBronze + medals.WinterBronze) * 1))

medals.sort_values('score', ascending=False).head(10)
```

# 27 สร้างคอลัมใหม่ชื่อ 'perGame' คำนวณคะแนนต่อการเข้าร่วมแข่งขัน (คะแนนคำนวณเหมือนเดิม แล้วหารด้วยจำนวนเกมที่เข้าแข่งแยกกันร้อนหนาว ค่านี้จะเป็นผลบวกของคะแนนฤดูร้อนและหนาว)

```python
medals['perGame'] = ((((medals.SummerGold * 3) + (medals.SummerSilver * 2) + (medals.SummerBronze * 1)) / medals.SummerGame)
                  + (((medals.WinterGold * 3) + (medals.WinterSilver * 2) + (medals.WinterBronze * 1)) / medals.WinterGame))

medals.sort_values('perGame').head(10)
```
